```python
def gap_dips_to_table(df): 
    for i in range(len(df)):
        dips_pair = [df[i]['% x_strip (m)'].iloc[0], df[i]['dip_mag'].iloc[0]]
        gaps_dips.append(dips_pair)
    gaps_dips = np.array(gaps_dips)

    return gaps_dips
```

```python
def dips_and_freqs(df):
    df['% x_strip (m)'] = df['% x_strip (m)'].round(6)  # to normalize float precision
    unique_vals = sorted(df['% x_strip (m)'].unique())
    dips_freqs = []
    for val in unique_vals:
        gap_d = df[df['% x_strip (m)']==val]
        min = gap_d[gap_d['S21 (1)']==np.min(gap_d['S21 (1)'])]
        dips_freqs.append([min['freq (GHz)'], min['S21 (1)']])
    dips_freq = np.array(np.hstack(dips_freqs))
    return dips_freq
```

```python
unique_vals = sorted(LN2_p2p3_2e5step['% x_strip (m)'].unique())
unique_vals
```

```python
dipfreqs = dips_and_freqs(LN2_p2p3_2e5step)
```
